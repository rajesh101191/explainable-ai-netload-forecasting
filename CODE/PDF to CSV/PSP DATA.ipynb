```python
import pdfplumber
import pandas as pd
import glob
import re
from tqdm import tqdm
import os

# ===============================
# Folder Path (UPDATED AS GIVEN)
# ===============================
pdf_folder = r"C:\Users\rjesh\PSP DATA\*.pdf"

output_file = r"C:\Users\rjesh\PSP DATA\State_Power_Position_All_Days.csv"

# Remove output file if already exists
if os.path.exists(output_file):
    os.remove(output_file)

print("Starting batch extraction...\n")

for file in tqdm(glob.glob(pdf_folder)):
    
    # ------------------------------
    # Extract Date from File Name
    # Example: 01.01.14_NLDC_PSP.pdf
    # ------------------------------
    filename = os.path.basename(file)
    date_match = re.search(r'\d{2}\.\d{2}\.\d{2}', filename)
    
    if date_match:
        report_date = date_match.group()
    else:
        report_date = "Unknown"
    
    with pdfplumber.open(file) as pdf:
        
        for page in pdf.pages:
            tables = page.extract_tables()
            
            for table in tables:
                df = pd.DataFrame(table)
                
                # Identify STATE TABLE using keyword
                if df.astype(str).apply(lambda x: x.str.contains("Punjab").any()).any():
                    
                    # Set header row
                    df.columns = df.iloc[0]
                    df = df[1:]
                    
                    # Drop completely empty rows
                    df = df.dropna(how="all")
                    
                    # Add Date column
                    df["Date"] = report_date
                    
                    # Write incrementally (no memory overload)
                    df.to_csv(output_file, 
                              mode='a', 
                              header=not os.path.exists(output_file), 
                              index=False)

print("\n✅ DONE! CSV Created Successfully.")
print("File saved at:", output_file)
```
